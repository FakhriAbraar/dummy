```python
import feedparser

# URL RSS Google Trends Indonesia
url = "https://trends.google.com/trending/rss?geo=ID"

# Mengambil data (Parsing)
feed = feedparser.parse(url)

print(f"--- {feed.feed.title} ---\n")

for entry in feed.entries:
    print(f"Topik: {entry.title}")
    print(f"Estimasi Pencarian: {entry.ht_approx_traffic}")
    if entry.get('ht_news_item_url'):
        print(f"Link Berita: {entry.ht_news_item_url}")
    print(f"Published: {entry.published}")
    print("-" * 20)
```

```python
import json

# Mengambil metadata utama dari feed
feed_metadata = feed.entries

# Print dengan indentasi 4 spasi agar rapi
print(json.dumps(feed_metadata, indent=4, default=str))

nama_file = "trends24_indonesia.json"
with open(nama_file, "w", encoding="utf-8") as f:
    json.dump(feed_metadata, f, indent=4, ensure_ascii=False)
```
